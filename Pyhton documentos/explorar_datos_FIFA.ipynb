### Estadísticos de variable de interés

La variable de estudio es el rating de las películas. 

Cambiar variable de interés a Age (Edad).




```python
print("Age")
print("Valor mínimo:", datos['Age'].min())
print("Valor máximo:",datos['Age'].max())
print("La media aritmética:",datos['Age'].mean())
print("Desviación Estándar:",datos['Age'].std())
print("Núm. de observaciones:",datos['Age'].count())
```

### Nacionalidad



```python
nacionalidad = pd.Categorical(datos['Nationality'])
nacionalidad.describe()
```

```python
frecuencia_nac = (datos.groupby("Nationality").agg(frequency=("Nationality", "count")))
type(frecuencia_nac)
# frecuencia_nac
frecuencia_nac = frecuencia_nac.sort_values(['frequency'], ascending=False)
frecuencia_nac_topten = frecuencia_nac.head(10)
frecuencia_nac_topten

```

```python
frecuencia_nac_topten.plot(kind="bar")
```

```python
pie_preferido = pd.Categorical(datos['Preferred.Foot'])
pie_preferido.describe()
```

### Visualizar datos

La variable de interés es el idioma (foreign). Se determina la frecuencia.



```python
ceros = datos['foreign'] == 0
print(ceros.value_counts())


frecuencia = (datos.groupby("foreign").agg(frequency=("foreign", "count")))
frecuencia
```

```python
datos.groupby('foreign').count()['name'].plot(kind='bar')
```

La variable de interés es el género (genre). Se determina la frecuencia.

```python
datos.groupby('genre').count()['name'].plot(kind='bar')
```
